```python
from sqlalchemy import create_engine
from urllib.parse import quote  

h = confidential['host']

p = confidential['port']
u = confidential['user']
pwd = confidential['password']
db = 'preprocessed_post_cls'
print(pwd)
connect_string = f'mysql+pymysql://{u}:%s@{h}:{p}/{db}' % quote(pwd)

# print(connect_string)
engine = create_engine(connect_string)
```

```python
from pandas.io import sql

DATAPATH = '/root/autodl-tmp/20m_tag'

file_ls = os.listdir(DATAPATH)


with open('log', 'a') as f:
    f.write('')

with open('log', 'r') as f:
    upladed = f.read().split('\n')

file_ls = list(map(lambda x: os.path.join(DATAPATH, x), file_ls))

file_ls = list(set(file_ls).difference(upladed))
 
for data_path in tqdm(file_ls):
    if data_path[-6:]=='pickle':  
        with open(data_path, 'rb') as f:
            df = pickle.load(f)
            df.to_sql(
                con=engine, 
                name='20m_id_tag', 
                if_exists='append', 
                index = False,
                method = 'multi',
                #chunksize = 5000,
                # dtype = {
                #     'id': "string",
                #     'occasion_cn': "string",
                #     'need_cn': "string",
                #     'who_cn': "string"
                # }
           )
        with open('log', 'a') as f:
            f.write('\n'+data_path)

    # break
```
